## Understanding how many data is missing

### Using a loop for counting the missing data

```python
missing_data = accident.isnull()
for faltantes in missing_data.columns.values.tolist():
    print(faltantes)
    print(missing_data[faltantes].value_counts())
    print("")
```

```python
missing_data = accident.isnull().sum(axis=0).reset_index()
missing_data.columns = ['column_name','missing_count']
missing_data = missing_data.sort_values(by='missing_count')

ind = np.arange(missing_data.shape[0])
width = 1
fig,ax = plt.subplots(figsize=(10,13))
rects = ax.barh(ind,missing_data.missing_count.values,color='red')
ax.set_yticks(ind)
ax.set_yticklabels(missing_data.column_name.values, rotation='horizontal')
ax.set_xlabel("Missing Values")
ax.set_title("Missing Data")
plt.show()
```


## Dropping the missing values

Here I had to analyze wether it was a good idea to drop the entire columns or just the rows. For the EXCEPTRSNCODE and the EXCEPTRSNDESC columns the values missing were too high and the meaning of them too low so drop the entire column was a good idea. The INTKEY column can not be fulfilled by data of the context and the values missing were too high as well, so I had to drop the entire column. For the INATTENTIONIND there were too little information so dropping the entire column as well does not have any influence. For the PEDROWNOTGRNT there were just 4667 entries so it was ok to drop the column. The SDOTCOLNUM column is arbitrary and does not respond to any question towards the severity of the accident so it has been dropped. The SPEEDING column had just 9333 entries so in the big picture it was not a good indicator, even thought it is supposed to be an accurate indicator of the severity of the enjuries. The column REPORTNO does not give any important information for the assignment. The column STATUS does not give any important information neither. 

For the other missing values, I dropped just the missing values (in rows) and not the entire column.

```python
accident.drop(["EXCEPTRSNCODE"], axis=1, inplace=True)
accident.drop(["EXCEPTRSNDESC"], axis=1, inplace=True)
accident.drop(["INTKEY"], axis=1, inplace=True)
accident.drop(["INATTENTIONIND"], axis=1, inplace=True)
accident.drop(["PEDROWNOTGRNT"], axis=1, inplace=True)
accident.drop(["SDOTCOLNUM"], axis=1, inplace=True)
accident.drop(["SPEEDING"], axis=1, inplace=True)
accident.drop(["INCKEY"], axis=1, inplace=True)
accident.drop(["COLDETKEY"], axis=1, inplace=True)
accident.drop(["REPORTNO"], axis=1, inplace=True)
accident.drop(["STATUS"], axis=1, inplace=True)
accident.drop(["OBJECTID"], axis=1, inplace=True)

accident.dropna(subset=["ST_COLCODE"], axis=0, inplace=True)
accident.dropna(subset=["ST_COLDESC"], axis=0, inplace=True)
accident.dropna(subset=["WEATHER"], axis=0, inplace= True)
accident.dropna(subset=["ROADCOND"], axis=0, inplace=True)
accident.dropna(subset=["LIGHTCOND"], axis = 0, inplace = True)
accident.dropna(subset=["UNDERINFL"], axis=0, inplace=True)
accident.dropna(subset=["JUNCTIONTYPE"], axis=0, inplace=True)
accident.dropna(subset=["COLLISIONTYPE"], axis=0, inplace=True)
accident.dropna(subset=["LOCATION"], axis=0, inplace=True)
accident.dropna(subset=["X"], axis=0, inplace=True)
accident.dropna(subset=["Y"], axis=0, inplace=True)
accident.dropna(subset=["ADDRTYPE"], axis=0, inplace=True)

accident.reset_index(drop=True, inplace=True)
```
